Importação das Bibliotecas

```python
import pandas as pd
#importar a biblioteca "pandas" na qual é utilizada para criação a expressão "as" simbolizando a abreviação da palavra por "pd= pandas".

import plotly.express as px
#importar a biblioteca "plotly.express" na qual é utilizada para criação de gráficos, a expressão "as" simbolizando a abreviação da palavra por "px= plotly.express".

import smtplib
#biblioteca de envio de e-mail

import email.message as em
#importar a biblioteca "email.message" na qual é utilizada para a expressão "as" simbolizando a abreviação da palavra por "em= email.message"
```

**Passo 1** - *Importar a base de dados (Pasta Vendas)*

```python
tb_vendas = pd.read_excel("/content/drive/MyDrive/Python para Ciências de Dados/SOZINHA/Vendas+-+Base+de+Dados.xlsx")

display(tb_vendas)
```

**Passo 2** - *Calcular o produto mais vendido (em quantidade)*.

```python
tb_quantidade_produto = tb_vendas.groupby('Produto').sum()
tb_quantidade_produto = tb_quantidade_produto[['Quantidade']]
tb_quantidade_produto = tb_quantidade_produto.sort_values(by='Produto', ascending=True)
display(tb_quantidade_produto)
```

**Passo 3** - *Calcular o produto mais vendido (em faturamento*).

```python
tb_vendas['Faturamento'] = tb_vendas['Quantidade'] * tb_vendas['Valor Unitário']
tb_faturamento_produto = tb_vendas.groupby('Produto').sum()
tb_faturamento_produto = tb_faturamento_produto[['Faturamento']].sort_values(by='Faturamento', ascending=False)

display(tb_faturamento_produto)
```

**Passo 4** - *Calcular a loja/ Estado que mais vendeu (em faturamento)*

```python
tb_vendas['Faturamento'] = tb_vendas['Quantidade'] * tb_vendas['Valor Unitário']
tb_faturamento_estado = tb_vendas.groupby('Loja').sum()
tb_faturamento_estado = tb_faturamento_estado[['Faturamento']].sort_values(by='Faturamento', ascending=False)

display(tb_faturamento_estado)
```

**Passo 5** - *Calcular o ticket médio por loja/ Estado.*

```python

tb_vendas['Ticket Médio'] = tb_vendas['Valor Unitário']
tb_media_vendas = tb_vendas.groupby('Loja').mean(numeric_only=True)
tb_media_vendas = tb_media_vendas[['Ticket Médio']]
tb_media_vendas = tb_media_vendas.sort_values(by= 'Ticket Médio', ascending=False)#"tb_media_vendas" recebe ela mesma/ ".sort_values"= comando de ordenação (maior/ menor)/ "(by= 'TM'(ordenar pela coluna Ticket Médio) ", ascending= False" (False= do maior para o menor/ True= do menor para o maior))"
display(tb_media_vendas)
```

**Passo 6** - *Criar um gráfico/ dashboard da loja/ Estado que mais vendeu (em faturamento).*

```python
grafico = px.bar(tb_faturamento_estado, y='Faturamento', x=tb_faturamento_estado.index)
#"grafico= variável definida para o gráfico"
#px.bar= chamando a bilbioteca "px" e a expressão ".bar" simboliza o gráfico de "barras"
#dentro do parênteses é informada a tabela base de dados para o gráfico, na qual é definido como eixo "y= a coluna Faturamento" e o eixo "x= a tabela faturamento representado a coluna loja/ Estado, pois foi reconhecido como índice pelo pandas, neste caso é necessário a inclusão do .index"

grafico.show()
#função que apresenta o gráfico
```

```python
preco = float(input('Digite o valor: ')) #solicitando para que o user digite um valor para o cálculo
calcular_imposto2 = lambda preco: preco * 0.2
#def calcular_imposto(preco): "EQUIVALE A --->" calcular_imposto2 = lambda preco:
#return preco * 0.20  "EQUIVALE A --->" preco * 0.2

print (calcular_imposto2(preco))
```

```python
precos = [100, 899, 990, 110]

#preco_produtos = float(input('Digite os valores: ', ))
#precos = preco_produtos.append(precos)

impostos = list(map(calcular_imposto2, precos)) #"var" recebe a lista('list()') que vem da função('map()') / dentro da função MAP informar "a função" e "a lista" como parâmetros separados por ","
#"map" = dentro da funçaõ "map" deve-se inserir a FUNÇÃO a ser utilizada juntamente com a lista que pretende calcular separados por "vígrula"

print(impostos) #exibe a variável que recebe a lista juntamente com o map
```

```python
impostos = list(map(lambda x: x * 0.2, precos)) #função que recebe o valor(preço) e este valor retorna o valor multiplicado por 0.2
#"x" mantém sempre na função representando cada elemento da lista automaticamente

print(impostos) #exibe a variável que recebe a lista juntamente com o map
```

```python
from babel.numbers import format_currency #PARA IMPORTAR SOMENTE UMA FERRAMENTA DA BIBLIOTECA ("FROM" "NOME_DA_BIBLIOTECA" "IMPORT" "FORMAT_CURRENCY")

#"babel.numbers" - biblioteca que padroniza no real brasileiro, porém, possui os padrões de moedas de várias moedas do mundo.
#"format.currency" - para realizar a importação de uma única ferramenta sem a necessidade de importar todas as infos da biblioteca.

valor = 1000 #variável
valor_formatado = format_currency(valor, 'BRL', locale='pt_BR') #"FORMAT_CURRENCY("VARIÁVEL" , " 'BRL'- moeda brasileira " , "LOCALE= 'pt_BR' - localização= pt - português_BR - Brasil ")"
#variável que irá receber o valor formatado

print(valor_formatado)
```
